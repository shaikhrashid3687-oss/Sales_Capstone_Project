```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("D:\Sales_Capstone_Project\python_notebook\data_cleaning.ipynb\Superstore_Cleaned.csv")

# Sales trend
df.groupby("Order Year")["Sales"].sum().plot(kind="line", title="Sales Trend")
plt.show()

# Profit trend
df.groupby("Order Year")["Profit"].sum().plot(kind="line", title="Profit Trend")
plt.show()

# Sales by region
df.groupby("Region")["Sales"].sum().plot(kind="bar", title="Sales by Region")
plt.show()

# Profit by category
df.groupby("Category")["Profit"].sum().plot(kind="bar", title="Profit by Category")
plt.show()

# Discount vs Profit
plt.scatter(df["Discount"], df["Profit"])
plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show()
df.to_csv("D:\Sales_Capstone_Project\python_notebook\sales_analysis.ipynb", index=False)
```
